```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the Excel file
file_path = '/home/sahaj432/Desktop/code_organized/Clutering_Project/final_code_cluster/new_result.xlsx'
data = pd.read_excel(file_path)

# Define the metrics and methods for comparison
metrics_corrected = ["Adjusted Rand", "Ajusted Mutual", "V_measure", "Jaccard", "Acc", "Time"]
methods_corrected = ["Kmeans", "Our", "Aggl", "HDBSCAN", "DBSCAN", "FCM", "ISOSplit"]

# Prepare the data for plotting
def prepare_plot_data_adjusted(data, method, metrics):
    method_data = data[[f"{method} {metric}" for metric in metrics]]
    method_data.columns = metrics
    method_data['Method'] = method
    method_data['Dataset'] = data['Dataset']
    return method_data

# Concatenate data for all methods
all_data = [prepare_plot_data_adjusted(data, method, metrics_corrected) for method in methods_corrected]
concatenated_data = pd.concat(all_data)

# Melt the data to long format for plotting
long_format_data = concatenated_data.melt(id_vars=['Method', 'Dataset'], var_name='Metric', value_name='Value')

# Custom color palette
colors = sns.color_palette("tab10", len(methods_corrected))

# Setting up the figure for enhanced line plots
fig, axes = plt.subplots(len(metrics_corrected), figsize=(15, 5 * len(metrics_corrected)))

for i, metric in enumerate(metrics_corrected):
    metric_data = long_format_data[long_format_data['Metric'] == metric]
    mean_data = metric_data.groupby(['Dataset', 'Method']).mean().reset_index()
    std_data = metric_data.groupby(['Dataset', 'Method']).std().reset_index()

    # Plotting each method with customizations
    for j, method in enumerate(methods_corrected):
        method_mean = mean_data[mean_data['Method'] == method]
        method_std = std_data[std_data['Method'] == method]
        axes[i].errorbar(method_mean['Dataset'], method_mean['Value'], yerr=method_std['Value'], 
                         label=method, fmt='-o', capsize=5, color=colors[j])

    # Customizing the plot aesthetics
    axes[i].set_title(f'Performance of Clustering Methods for {metric}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Dataset', fontsize=12)
    axes[i].set_ylabel(metric, fontsize=12)
    axes[i].grid(True, which='both', linestyle='--', linewidth=0.5)
    axes[i].tick_params(axis='x', rotation=45)

# Placing the legend outside the plot
plt.legend(methods_corrected, loc='upper center', bbox_to_anchor=(-0.1, -0.15), fancybox=True, shadow=True, ncol=len(methods_corrected))

plt.tight_layout()
plt.show()
```
