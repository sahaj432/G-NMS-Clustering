```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
df = pd.read_excel('/home/sahaj432/Desktop/code_organized/Clutering_Project/final_code_cluster/new_result.xlsx')  # Replace with your file path

# Metrics and methods
metrics_corrected = ["Adjusted Rand", "Ajusted Mutual", "V_measure", "Jaccard", "Acc", "Time"]
methods_corrected = ["Kmeans", "Our", "Aggl", "HDBSCAN", "DBSCAN", "FCM", "ISOSplit"]

# Melting the DataFrame for easier plotting
melted_df = pd.melt(df, id_vars='Dataset', 
                    value_vars=[f"{method} {metric}" for method in methods_corrected for metric in metrics_corrected],
                    var_name='Method_Metric', value_name='Value')
```

```python
selected_metric = "Acc"  # Example metric
selected_NAME= "Acurracy (%)"
# Filtering data for the selected metric
filtered_data = melted_df[melted_df['Method_Metric'].str.contains(selected_metric)]


sns.set(style="whitegrid")
plt.figure(figsize=(15, 8), dpi=800)

# Create bar plot
sns.barplot(x='Dataset', y='Value', hue='Method_Metric', data=filtered_data, palette='viridis')

# Enhancing the plot
plt.xlabel('Experiments', fontsize=18, weight='bold')
plt.ylabel(selected_NAME, fontsize=18, weight='bold')
plt.xticks(rotation=90, fontsize=18, weight='bold')
plt.yticks(fontsize=18, weight='bold')

# Getting the current handles and labels from the plot
handles, labels = plt.gca().get_legend_handles_labels()

# Custom names for the legend
custom_legend_names = ["K-Means", "G-NMS*", "Agglomerative", "HDBSCAN", "DBSCAN", "FCM", "ISO-SPLIT"]

# Ensure that there are enough custom labels for the handles
if len(handles) == len(custom_legend_names):
    legend = plt.legend(handles, custom_legend_names, title='Methods', loc='upper left', bbox_to_anchor=(1, 1), fontsize=18)
    plt.setp(legend.get_texts(), fontsize='18', weight='bold') # for legend text
    plt.setp(legend.get_title(), fontsize='18', weight='bold')
else:
    print("Warning: The number of custom legend names does not match the number of legend items.")

# Adjust layout and save in high resolution
plt.tight_layout()
plt.savefig('result_Acc.png', dpi=800)

plt.show()
```
