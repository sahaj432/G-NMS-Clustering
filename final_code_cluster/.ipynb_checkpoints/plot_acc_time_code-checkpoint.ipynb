```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to the data file
excel_path = '/home/sahaj432/Desktop/code_organized/Clutering_Project/final_code_cluster/cluster_acc_time_result.xlsx'

# Read the data from the Excel file
data = pd.read_excel(excel_path)

# List of methods and their generic names for the legend
methods = ['Kmeans', 'Our', 'Aggl', 'HDBSCAN', 'DBSCAN', 'FCM', 'ISOSplit']
generic_method_names = ["K-Means", "G-NMS*", "Agglomerative", "HDBSCAN", "DBSCAN", "FCM", "ISO-SPLIT"]

# Calculate the average accuracy and average time for each method
average_acc = data[[f'{method} Acc' for method in methods]].mean()
average_time = data[[f'{method} Time' for method in methods]].mean()

# Different colors for each method
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'orange', 'black']

# Create the plot
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
for i, method in enumerate(methods):
    plt.scatter(
        average_time[i], 
        average_acc[i], 
        label=f'{generic_method_names[i]} (Acc: {average_acc[i]:.2f}, Time: {average_time[i]:.2f}s)',
        color=colors[i], 
        s=100  # Size of the marker
    )

# Setting the title and labels with font size and weight
plt.xlabel('Average Time (s)', fontsize=18, fontweight='bold')
plt.ylabel('Average Accuracy (%)', fontsize=18, fontweight='bold')

# Setting the tick parameters
plt.xticks(fontsize=18, fontweight='bold')
plt.yticks(fontsize=18, fontweight='bold')

# Adding the legend with font size and weight
legend = plt.legend(title='Methods', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=18, prop={'size': 18, 'weight': 'bold'})
plt.setp(legend.get_texts(), fontsize='18', weight='bold') # for legend text
plt.setp(legend.get_title(), fontsize='18', weight='bold')

# Adding the grid to the plot
plt.grid(True)

# Adjust the subplot parameters to give the plot more room
plt.subplots_adjust(right=0.7)

# Save the plot with adjusted bounding box
#plt.savefig('result_acc_time.png', dpi=800, bbox_extra_artists=(legend,), bbox_inches='tight')

# Display the plot
plt.show()
```
